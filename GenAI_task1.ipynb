**Task 1: Conceptual Understanding (Mandatory – Written)**

```python
# ==============================
# 📌 Install Required Libraries
# ==============================
!pip install nltk

# ==============================
# 📌 Import Libraries
# ==============================
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import string

# Download required datasets
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Added to resolve LookupError

# ==============================
# 📌 Sample Messy Text Data
# ==============================
text = """
OMG!!! I LOVED this movie soooo much 😍😍!!!
The acting was GREAT, but the storyline was kinda boring...
I wouldn't say it's the BEST movie, but it's not bad either!!!
"""

print("Original Text:\n", text)

# ==============================
# 📌 Step 1: Lowercasing
# ==============================
text_lower = text.lower()
print("\nLowercase Text:\n", text_lower)

# ==============================
# 📌 Step 2: Tokenization
# ==============================
tokens = word_tokenize(text_lower)
print("\nTokens:\n", tokens)

# ==============================
# 📌 Step 3: Remove Punctuation
# ==============================
tokens_no_punct = [word for word in tokens if word not in string.punctuation]
print("\nTokens without punctuation:\n", tokens_no_punct)

# ==============================
# 📌 Step 4: Remove Stopwords
# ==============================
stop_words = set(stopwords.words('english'))
tokens_no_stopwords = [word for word in tokens_no_punct if word not in stop_words]
print("\nTokens without stopwords:\n", tokens_no_stopwords)

# ==============================
# 📌 Step 5: Stemming
# ==============================
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens_no_stopwords]
print("\nStemmed Words:\n", stemmed_words)

# ==============================
# 📌 Step 6: Lemmatization
# ==============================
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens_no_stopwords]
print("\nLemmatized Words:\n", lemmatized_words)

# ==============================
# 📌 Final Clean Tokens
# ==============================
print("\nFinal Processed Tokens:\n", lemmatized_words)
```

**Task 2: Advanced Preprocessing Function**

```python
# Install & Import
!pip install nltk

import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

# Function
def preprocess_text(text):
    try:
        if not text or text.strip() == "":
            return [], ""

        text = text.lower()
        text = re.sub(r"http\S+|www\S+", "", text)
        text = re.sub(r"\S+@\S+", "", text)
        text = re.sub(r"\d+", "", text)
        text = re.sub(r"(.)\1{2,}", r"\1", text)
        text = re.sub(r"[^a-zA-Z\s]", "", text)
        text = re.sub(r"\s+", " ", text).strip()

        tokens = word_tokenize(text)

        tokens = [w for w in tokens if (len(w) > 2 or w in ["no", "not"])]
        tokens = [w for w in tokens if w not in stop_words]

        clean_sentence = " ".join(tokens)

        return tokens, clean_sentence

    except:
        return [], ""
```

**Task 3: Stress Testing**

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []
clean_sentences = []

for text in sample_inputs:
    tokens, clean_sent = preprocess_text(text)

    all_tokens.extend(tokens)
    clean_sentences.append(clean_sent)

    print("Original Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_sent)
    print("-" * 50)
```

**Task 4: Token Analytics**

```python
for i, text in enumerate(sample_inputs):
    tokens, _ = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    print(f"Sentence {i+1}:")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_len, 2))
    print("-" * 40)
```

**Task 5: Frequency Analysis**

```python
from collections import Counter

counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
least_common = sorted(counter.items(), key=lambda x: x[1])[:5]
print(least_common)
```

**Task 6: Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sent = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sent)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Test
output = full_pipeline(sample_inputs)
print(output)
```

**Task 7: Error Handling**

```python
edge_cases = [
    "",
    "😂😂😂😂",
    "1234567890"
]

for case in edge_cases:
    tokens, clean_sent = preprocess_text(case)

    print("Input:", case)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean_sent)
    print("-" * 40)
```
