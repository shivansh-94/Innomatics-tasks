Problem 1: Unique Words in a Sentence

```python
# Taking input from user
sentence = input("Enter sentence: ")

# Splitting sentence into words
words = sentence.split()

# Converting list to set to remove duplicates
unique_words = set(words)

# Printing results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Problem 4: Count Products Above Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

attendance_percentage = (present_count / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

Problem 6: Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set
unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

Problem 7: Count Character Frequency

```python
text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```

Problem 9: Check if Key Exists in Dictionary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000
}

key_to_check = "Ravi"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Problem 10: Calculate Average Marks

```python
marks = [70, 75, 80, 65]

total_marks = sum(marks)
number_of_subjects = len(marks)

average = total_marks / number_of_subjects

print("Average Marks:", average)
```
