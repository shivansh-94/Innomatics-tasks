**Problem 1: Smart Parking Lot Management System**

```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)

    if parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

**Problem 2: Online Food Delivery Time Estimator**

```python
def delivery_time_estimator(distance, traffic, weather):
    # Base time: 5 minutes per km
    base_time = distance * 5

    # Traffic delay
    if traffic == "Low":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    else:  # High
        traffic_delay = 20

    # Weather delay
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 5
    else:  # Rainy
        weather_delay = 15

    total_time = base_time + traffic_delay + weather_delay

    print("Estimated Delivery Time:", total_time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

delivery_time_estimator(distance, traffic, weather)
```

**Problem 3: Movie Theatre Seat Occupancy Analyzer**

```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy, 2), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150  # 150 seats booked

seat_occupancy_analyzer(total_seats, booked_seats)
```

**Problem 4: Cloud Server Load Classification System**

```python
def server_load_classifier(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", round(average, 2), "%")

    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

server_load_classifier(cpu_readings)
```

**Problem 5: Smart Classroom Resource Usage Monitor**

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    # Threshold: usage > 8 hours considered overuse
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_usage)
```

**Problem 6: Online Event Registration Capacity Controller**

```python
def event_registration_controller(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
