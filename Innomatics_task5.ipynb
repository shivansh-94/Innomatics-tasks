**Problem 1: Social Media – Post Engagement Analyzer**

```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# Example
likes = [200, 150, 300, 400]
analyze_post_engagement(likes)
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# Example
rainfall = [70, 75, 80, 65]
rainfall_checker(rainfall, 70)
```

**Problem 4: Social Media – Duplicate Account Detection**

```python
def detect_duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

# Example
users = ["shiv", "raj", "amit", "shiv"]
detect_duplicate_accounts(users)
```

**Problem 6: Agriculture – Premium Crop Price Filter**

```python
def premium_crop_filter(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

# Example
crop_prices = [1500, 2500, 1800, 3200]
premium_crop_filter(crop_prices)
```

**Problem 7: System Monitoring – Application Health Checker**

```python
def application_health_checker(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

# Example
application_health_checker(7)
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for day in attendance_list:
        if day == 1:  # 1 = Present
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

# Example
attendance = [1,1,1,0,1,1,1,1,0,1]
attendance_eligibility(attendance)
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
def electricity_bill_calculator(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Units Consumed:", units)
    print("Total Bill:", bill)
    print("Usage Classification:", usage)

# Example
electricity_bill_calculator(250)
```
