**Problem 1: Mobile Recharge Validation System**

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return "Recharge amount must be at least ‚Çπ50"
    elif amount not in valid_plans:
        return "Invalid plan selected. Available plans: 199, 299, 399, 599"
    else:
        return "Recharge Successful ‚úÖ"

while True:
    try:
        amt = int(input("Enter recharge amount: ‚Çπ"))
        result = validate_recharge(amt)
        print(result)

        if result == "Recharge Successful ‚úÖ":
            break
        else:
            print("Please try again.\n")

    except ValueError:
        print("Invalid input! Please enter a valid number.\n")
```

**Problem 2: Inventory Reorder Alert System**

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert ‚ö†Ô∏è (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK ‚úÖ (Stock = {quantity})")

inventory = {
    "Apples": 20,
    "Bananas": 10,
    "Oranges": 5,
    "Mangoes": 25
}

check_inventory(inventory)
```

**Problem 3: Student Result Processing System**

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average = {average:.2f} ‚Üí Pass ‚úÖ"
    else:
        return f"Average = {average:.2f} ‚Üí Fail ‚ùå"

marks_list = list(map(int, input("Enter marks separated by space: ").split()))
result = process_result(marks_list)
print(result)
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare

while True:
    try:
        dist = float(input("Enter distance (in km): "))
        peak_hour = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(dist, peak_hour)
        print(f"Total Fare = ‚Çπ{total_fare:.2f} üöñ")

        break

    except ValueError:
        print("Invalid input! Please try again.\n")
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return f"Attendance = {percentage:.2f}% ‚Üí Eligible ‚úÖ"
    else:
        return f"Attendance = {percentage:.2f}% ‚Üí Not Eligible ‚ùå"

attendance = input("Enter attendance (P/A separated by space): ").split()
result = check_attendance(attendance)
print(result)
```

**Problem 6: Password Strength Checker**

```python
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password ‚ùå (Minimum length 8 required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password ‚úÖ"
    else:
        return "Weak Password ‚ùå (Must contain at least one digit and one special character @#$)"

pwd = input("Enter password: ")
print(check_password(pwd))
```
